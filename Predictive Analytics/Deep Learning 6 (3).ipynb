```python
from __future__ import print_function

import keras
#from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, LeakyReLU, Merge, concatenate, Maximum
from keras.optimizers import SGD
from keras import regularizers

import numpy as np
import pickle
np.random.seed(0) 
import warnings 
warnings.filterwarnings("ignore")

batch_size = 128
num_classes = 3

```

```python
N = 100 # number of points per class
D = 2 # dimensionality
K = 3 # number of classes
X = np.zeros((N*K,D))
y = np.zeros(N*K, dtype='uint8')
for j in range(K):
    ix = range(N*j,N*(j+1))
    r = np.linspace(0.0,1,N) # radius
    t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # theta
    X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
    y[ix] = j

pickle.dump(X,open('dataX.pickle','wb'))
pickle.dump(y,open('dataY.pickle','wb'))
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
print(x_train.shape, 'train samples')
print(x_test.shape[0], 'test samples')


# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
epochs=200
model = Sequential()
model.add(Dense(100, activation='linear',kernel_regularizer=regularizers.l2(1e-3), input_shape=(2,)))
model.add(Dense(num_classes,activation='softmax'))

sgd = SGD(lr=1e-0, decay=1e-6, momentum=0.9, nesterov=True)

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

# Q5 (3)

```python
epochs=3000
model1 = Sequential()
model1.add(Dense(100, activation='linear',kernel_regularizer=regularizers.l2(1e-4), input_shape=(2,)))
model2 = Sequential()
model2.add(Dense(100, activation='linear',kernel_regularizer=regularizers.l2(1e-4), input_shape=(2,)))

merged_model = Sequential()
merged_model.add(Merge([model1, model2], mode='max', concat_axis=1))
merged_model.add(Dense(num_classes, activation='softmax',kernel_regularizer=regularizers.l2(1e-4)))

sgd = SGD(lr=0.1)

merged_model.summary()

merged_model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])

history = merged_model.fit(x=[x_train,x_train], y=y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=([x_test,x_test], [y_test]))
score = merged_model.evaluate([x_test,x_test], y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
